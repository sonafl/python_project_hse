```python
# Импортирум модуль random, который понадобится при замене вопросов. 
import random 

# Создаем три словаря для трех уровней сложности,
# ключ - вопрос, значение - ответ.  
easylevel = {'Вид тропа, в основе которого — перенос свойств одних предметов и явлений на другие («пламя таланта»), обозначается термином ___________': 'МЕТАФОРА', 
           'К какому направлению литературы принадлежит «Герой нашего времени»?':'РЕАЛИЗМ', 
           'Как называется речь, далекая от литературной формы?': 'ПРОСТОРЕЧИЕ',
           'Как называется резкое противопоставление предметов и явлений в художественном произведении?':'АНТИТЕЗА', 
           'Укажите авторское определение жанра «Евгения Онегина».':'РОМАН В СТИХАХ', 
           'Как называется созвучие концов стихотворных слов (час — пронеслась; ненастной — ужасный и т. п.)':'РИФМА', 
           'Как называются вопросы, не требующие ответа? Укажите только ПРИЛАГАТЕЛЬНОЕ В ИМ. ПАДЕЖЕ.':'РИТОРИЧЕСКИЕ',
           'К какому литературному роду относятся стихотворения? Укажите СУЩЕСТВИТЕЛЬНОЕ В ИМ ПАДЕЖЕ.':'ЛИРИКА', 
           'Как называется художественное определение?':'ЭПИТЕТ', 
           'Назовите литературное направление, у истоков которого стоял В. А. Жуковский и принципы которого нашли свое воплощение в данном произведении.':'РОМАНТИЗМ'}

intermlevel = {'Укажите термин, которым обозначают повтор слова или группы слов в начале соседних фраз.':'АНАФОРА',
             'Укажите вид рифмовки (АВАВ).':'ПЕРЕКРЁСТНАЯ',
             'Как называется такая форма речевого общения, где чередуются обращённые друг к другу высказывания героев?':'ДИАЛОГ',
             'Поэтическое течение, в русле которого развивалось творчество А.А. Блока, это _______':'СИМВОЛИЗМ',
             'Как называется выразительная подробность в художественном тексте?':'ДЕТАЛЬ',
             'Каким термином обозначается описание природы в художественном произведении («а тут сияло солнце, вздыхала степь, блестела трава в брильянтах дождя и золотом сверкала река»)?':'ПЕЙЗАЖ',
             'К какому типу героя относится Данко из рассказа «Старуха Изергиль»? Укажите ПРИЛАГАТЕЛЬНОЕ В ИМ. ПАДЕЖЕ.':'РОМАНТИЧЕСКИЙ',
             'Непривычный порядок слов в стихотворении – это ____________ ':'ИНВЕРСИЯ',
             'Укажите род литературы, к которому относится произведение А. Т. Твардовского «Василий Тёркин». Укажите СУЩЕСТВИТЕЛЬНОЕ':'ЛИРОЭПОС',
             'Как в литературоведении называется столкновение характеров и обстоятельств, лежащее в основе развития действия?':'КОНФЛИКТ'}

hardlevel = {'Как называется трехсложный стихотворный размер, в котором первый слог УДАРНЫЙ?':'ДАКТИЛЬ',
           'К какому жанру принадлежит «Чистый понедельник» И.А. Бунина?':'РАССКАЗ',
           'Произведение «Гамлет» Б.Л. Пастернак «приписал» герою своего известного романа «________»':'ДОКТОР ЖИВАГО',
           'Укажите жанр, к которому можно отнести рассказ старухи Изергиль о Ларре?':'ЛЕГЕНДА', 
           'Каким термином обозначается отображение внутренних, душевных движений личности в литературном произведении?':'ПСИХОЛОГИЗМ',
           'Каким термином обозначается взаимосвязь и расположение частей в художественном произведении?':'КОМПОЗИЦИЯ',
           'Повторение строк в начале и в конце произведения свойственно композиции _______.':'ОБРАМЛЕНИЯ', 
           'Одним из основных средств характеристики героев в драматических произведениях является ______.':'РЕЧЬ',
           'Для «Недоросля» характерны _________ фамилии.':'ГОВОРЯЩИЕ',
           'Укажите род литературы, к которому относится произведение Н. А. Некрасова «Кому на Руси жить хорошо». Укажите СУЩЕСТВИТЕЛЬНОЕ.':'ЛИРОЭПОС'}
# Создаем список, состоящий из строк с советами. 
advice = ['Не будь Толстым - пиши сочинения кратко и на русском языке!', 'Не будь Есениным - никакого мата в сочинениях, какая бы тема не была!', 
        'Помни о структуре написания сочинения, никакой лесенки, как у Маяковского!', 'Во время экзамена забудь про пессимизм - не будь Лермонтовым!', 'Верь в успех, как Воланд верит в Бога!']

# Создаем два списка, состоящих из ключей словарей легкого и среднего уровней. 
## Это нужно для замены вопросов.  
elevel = list(easylevel.keys()) 
interlevel = list(intermlevel.keys())

# Функция describe_rules выводит правила прохождения теста  для пользователя. 
def describe_rules():
    print('Отлично! Давай разберем несколько правил:')
    print('1. Следи за тем, чтобы ответы грамматически соответствовали предложению или отвечали точно на поставленный вопрос.')
    print('2. Не забывай писать буквы Ё/ё, на экзамене это важно!')
    print('3. За правильные ответы ты получаешь 1 балл.')
    print('4. Если ты ответил неправильно, на некоторых уровнях ты можешь ответить на вопрос полегче и получить 0.5 баллов за правильный ответ.') 
    
# Функция sort_dict_by_value превращает словарь в список кортежей.
## С помощью этой функции можно отсортировать словарь по значениям.
def sort_dict_by_value(dictionary, reversed_order):
    list_d = list(dictionary.items())   # Преобразуем словарь в список.
    new_list = []                       # Создаем пустой список, в котором ключ и значение поменяются местами.
    for key, value in list_d:           # Последовательно для всех ключей и значений:  
        new_list.append((value, key))   # добавляем их в список в виде кортежа (значение, ключ).
    list_sorted = sorted(new_list, reverse=reversed_order) # Сортируем список кортежей по убыванию.
    return list_sorted                  # Возвращаем отсортированный список кортежей.
    
# Функция to_make_rating выводит рейтинг пользователей,
# использует работу с файлом. 
def to_make_rating(name, scores):
    print('Рейтинг пользователей:')
    # Создаем в пустой словарь, где ключом будут баллы,
    # а значением - имя пользователя.
    dict_for_rating = {}  
    # Добавлям данные из csv-файла в словарь.
    with open('rating.csv', encoding = 'UTF-8') as r:
        for i in r:
            a = i.strip('\n').split(';')
            dict_for_rating[a[0]] = float(a[1])
    # Выводим рейтинг пользователей, отсортировав словарь по убыванию.
    for item in sort_dict_by_value(dict_for_rating, True):
        print(*item)

# Функция show_results выводит результат и рандомный совет. 
def show_results(scores):
    print(f'Твой балл: {scores}')
    print(random.choice(advice))

# Функция start_easy_game выводит вопросы легкого уровня и считает баллы.
## Функция возвращает количество баллов. 
def start_easy_game(scores):
    for i in easylevel:
        answer = input(f'{i} ').strip().upper() #Спрашиваем у пользователя вопрос и ожидаем ответ. 
        if answer == easylevel[i]: # Проверяем, совпадает ли ответ пользователя со значением ключа. 
            print('Правильный ответ!') 
            scores += 1  # За правильные ответы начисляется один балл. 
        else:
            print('Ответ неправильный.')
    return scores

# Функция start_interm_game выводит вопросы среднего уровня, 
# заменяет в определенных случаях вопросы и считает баллы. 
## Принцип работы примерно схож с работой функции start_easy_game. 
def start_interm_game(scores):
    for i in intermlevel:
        answer = input(f'{i} ').strip().upper() 
        if answer == intermlevel[i]:
            print('Правильный ответ!')
            scores += 1  
        # Предложим пользователю ответить на более легкий вопрос. 
        ## Используем функцию choice из модуля random. 
        ### Замена вопросов будет при каждом неправильном ответе.
        else: 
            print('Давай попробуем вопрос полегче.')
            for i in range(1):
                q = random.choice(elevel)
                answer1 = input(f'{q} ').strip().upper()
                if answer1 == easylevel[q]:  
                    print('Правильный ответ!')
                    scores += 0.5 # За правильный ответ в зоне "замены вопросов" дается 0.5 баллов. 
                else:
                    print('Ответ неправильный. Попыток больше нет.')
    return(scores)

# Функция start_hard_game выводит вопросы трудного уровня, 
# заменяет в определенных случаях вопросы и считает баллы. 
## Принцип работы идентичен работе функции start_interm_game. 
def start_hard_game(scores, counter):
    for i in hardlevel:
        answer = input(f'{i} ').strip().upper()
        if answer == hardlevel[i]:
            print('Правильный ответ!')
            scores += 1
        else:
            print('Давай попробуем вопрос полегче.')
            for i in range(1):
                counter += 1
                q = random.choice(interlevel)
                answer1 = input(f'{q} ').strip().upper()
                if answer1 == intermlevel[q]: 
                    print('Правильный ответ!')
                    scores += 0.5
                else:
                    print('Ответ неправильный. Попыток больше нет.')
    return scores, counter

# ОСНОВНОЙ КОД

scores = 0 # В этой переменной будут считываяться баллы за правильные ответы. 
# Представляемся пользователю.
counter = 0
print('Привет, я Хина! Меня назвали в честь таксы известного писателя Антона Чехова. Я помогу тебе подготовиться к тестовой части ЕГЭ по литературе.')
# Начинаем тест и знакомим пользователя с правилами.
hello = input('Напиши "Начинаем", если хочешь начать прямо сейчас ').lower().strip()
if hello == 'начинаем':
    describe_rules()
    # Предлагаем пользователю выбрать уровень. 
    level = input('Выбери уровень сложности: лёгкий, средний или трудный. Напиши прилагательное соответствующего уровня: ').lower().strip()
    # Создаем цикл while, который работает пока не введут название уровня корректно.
    while level != 'легкий'and level != 'лёгкий' and level != 'средний' and level != 'трудный':
        level = input('Не поняла тебя. Введи название уровня еще раз ').lower().strip()
    # В зависимости от уровня задаются вопросы из соответствующего словаря.
    if level == 'лёгкий' or level == 'легкий':   
        scores = start_easy_game(scores)
        # После прохождения теста выводим баллы и рандомный совет.
        show_results(scores)
        print('До новых встреч!') # Прощаемся с пользователем.
    # Делаем то же самое при выборе других уровней. 
    elif level == 'средний':
        scores = start_interm_game(scores)
        show_results(scores)
        print('До новых встреч!') 
    # Совершаем со сложным уровнем те же действия, что и со средним. 
    elif level == 'трудный':
        scores = start_hard_game(scores, counter)             
        show_results(scores) 
        print(counter)
        # После прохождения только трудного уровня будет предложено быть в рейтинге. 
        ## Запрашиваем у пользователя имя для включения его в рейтинг
        ## или слово "нет", если пользователь этого не хочет.
        name = input('Напиши своё имя, если хочешь быть в рейтинге. Напиши "нет", если не хочешь. ').lower().strip()
        if name != 'нет':
           # Создаем csv-файл (в первый раз), 
           # дальше информация в файле будет редактироваться.
           ## В файле: баллы и имена пользователей. 
            with open('rating.csv','a', encoding='UTF-8') as r:
                print(name, scores, sep=';', file=r)
            # Выводим рейтинг пользователей.
            rating = to_make_rating(name, scores)
            print('До новых встреч!')
        else:
            # Если пользователь не хочет быть в рейтинге, 
            # выведем ему рейтинг других участников. 
            print('Хорошо, я тебя поняла.')
            rating = to_make_rating(name, scores)
            print('До новых встреч!')      
else:
    print('Не хочешь проходить тест? Тогда я жду тебя в другое время!')
```
