Я отправила вместе с кодом изображение, его нужно загрузить в Юпитер, чтобы оно сработало в коде, а также csv-файл с тестовым пользователем.:)

```python
# Импортирум модуль random, который понадобится при замене вопросов. 
import random 
# Используем модуль PIL для загрузки и показа изображения. 
from PIL import Image

# Создаем три словаря для трех уровней сложности,
# ключ - вопрос, значение - ответ.  
elementarylevel = {'Вставь на место пропуска ГЛАГОЛ в настоящем времени: «My sister and I ______ both blonde».': 'ARE', 
                '«That is the man _____ works in our company». На место пропуска нужно вставить WHO или WHAT? Впиши ВЕРНЫЙ ответ.':'WHO', 
                'Употреби глагол BE в правильной форме: «Where ____ you yesterday?»': 'WERE',
                'Вставь на место пропуска нужный ПРЕДЛОГ: «I often go out for a drink ____ Saturdays».':'ON', 
                'Какой ответ на вопрос «What is your best friend like?» ПРАВИЛЬНЫЙ? Укажи НОМЕР верного ответа.':'3', 
                'Напиши слово HALF в множественном числе.':'HALVES', 
                'Употреби глагол TEACH в правильной форме: «He______ English in Chine last year».':'TAUGHT',
                'Нужно ответить верно на вопрос: – Are you hungry?  – Yes, _______ . ':'I AM', 
                'Вставь на место пропуска нужный ПРЕДЛОГ: «I want to learn to play ______ guitar».':'ON', 
                'Верно ли, что вопрос «How do you do?» переводится как «Здравствуйте»? Укажи ДА или НЕТ.':'ДА'}

intermlevel = {'Употреби глагол ATTEMPT в правильной форме: «The man was accused of _____ a burglary».':'ATTEMPTING',
             'Что нужно вставить на место пропуска: «I am going to the hairdressers today. It’s time to______». Укажи НОМЕР верного ответа.':'4',
             'Вставь на место пропуска нужный ПРЕДЛОГ: «Their address must be ______ the website».':'ON',
             'Употреби глагол SUIT в правильной форме: «I think this dress _____  you. You look pretty».':'SUITS',
             'Вставь на место пропуска нужный ПРЕДЛОГ: «I’m sorry, I didn’t hear the alarm go______».':'OFF',
             'Употреби глагол FLY в правильной форме будущего времени: «This time tomorrow I ______ to Japan».':'WILL BE FLYING',
             '«There’s _______ point in complaining. They won’t listen to us». Что нужно вставить в пропуск: LITTLE или FEW? Впиши ВЕРНЫЙ ответ.':'LITTLE',
             'Что нужно вставить на место пропуска: «I_____ an umbrella. It did not rain after all». Укажи НОМЕР верного ответа.':'1',
             'Употреби глагол BE в правильной форме: «There ____ some cheese and ham left, I can make a sandwich».':'IS',
             'Употреби глагол BUY в правильной форме: « I ______ already a car».':'HAVE BOUGHT'}

advancedlevel = {'Употреби глагол TAKE в правильной форме: «The minister denied______ the bride».':'TAKING',
               'Что нужно вставить на место пропуска: «What_____ he is!». Укажи НОМЕР верного ответа.':'2',
               'Вставь на место пропуска нужный ПРЕДЛОГ: «I don’t think he is capable ____ climbing that mountain».':'OF',
               'Что нужно вставить на место пропуска: «It ________ to be rude». Укажи НОМЕР верного ответа.':'3', 
               '«The match was held  ______ the bad weather». Что нужно вставить на место пропуска: DESPITE или IN SPITE? Впиши верный ответ.':'DESPITE',
               'Употреби верный МОДАЛЬНЫЙ глагол: «I have no idea where she is. She ______ be waiting for us outside the pub»':'SHOULD',
               'Вставь на место пропуска нужный ПРЕДЛОГ: «John’s hard work paid _____ and he passed his exam easily».':'OFF', 
               'Употреби верный ГЛАГОЛ в предложении: «My manager has recently informed me that I will ______ redundant».':'GET',
               'Употреби глагол DETERMINE в правильной форме: «The girl we interviewed yesterday seems to be quite _______».':'DETERMINED',
               'Что нужно вставить на место пропуска: «There were ______ any people in the cinema yesterday».':'1'}

# Создаем два списка, состоящих из ключей словарей начального, среднего, продвинутого уровней. 
## Это нужно для замены вопросов и тестов с multiple choice, 
## так как для идентификации подобного вопроса, нужны будут индексы.  
alevel = list(elementarylevel.keys()) 
blevel = list(intermlevel.keys())
clevel = list(advancedlevel.keys())

# Создаем три списка, состоящих из ответов для вопросов с multiple choice.
a_level_answers = ['1. He likes skiing.', '2. He is tall and slim.', '3. He is cheerful and funny.', '4. He is like dogs.'] 
b_level_answers = ['1. cut my hair', '2. my hair be cut', '3. make my hair cut', '4. have my hair cut', "1. needn't have taken", "2. did not need to take", "3. did not have to take", "4. shouldn't have taken"]
c_level_answers = ['1. bore', '2. a bore', '3. bored', '4. boring', '1. is he doesn’t like', '2. is not him', '3. is not like him', '4. doesn’t like him', '1. hardly', '2. less', '3. few', '4. little']

# Используем декоратор sampleDecorator, который будет декорировать функцию show_results.
## Декоратор будет показывать изображение с надписью "барабанная дробь". 
def sampleDecorator(func):
    # Название функции условное. 
    def decorared_function(a, b, c):
        print("Ты готов?")
        img = Image.open('1.jpeg')
        # Показать изображение.
        img.show()
        func(a, b, c)
    return decorared_function

# Функция describe_rules выводит правила прохождения теста  для пользователя. 
def describe_rules():
    print('Отлично! Давай разберем несколько правил:')
    print('1. Тебе нужно выбрать один из трёх уровней: elementary (начальный), intermediate (средний) или advanced (продвинутый).')
    print('2. В тесте будут вопросы, когда тебе нужно дать краткий ответ или выбрать номер правильного варианта.')
    print('3. За правильные ответы ты получаешь 1 балл.')
    print('4. Если ты ответил неправильно, на некоторых уровнях ты можешь ответить на вопрос полегче и получить 0.5 баллов за правильный ответ. Замена вопросов поможет точнее определить твой уровень!')
    print('5. При завершении продвинутого уровня ты можешь попасть в общий рейтинг пользователей.')
    
# Функция sort_dict_by_value превращает словарь в список кортежей.
## С помощью этой функции можно отсортировать словарь по значениям.
def sort_dict_by_value(dictionary, reversed_order):
    list_d = list(dictionary.items())   # Преобразуем словарь в список.
    new_list = []                       # Создаем пустой список, в котором ключ и значение поменяются местами.
    for key, value in list_d:           # Последовательно для всех ключей и значений:  
        new_list.append((value, key))   # добавляем их в список в виде кортежа (значение, ключ).
    list_sorted = sorted(new_list, reverse=reversed_order) # Сортируем список кортежей по убыванию.
    return list_sorted                  # Возвращаем отсортированный список кортежей.
    
# Функция to_make_rating создаёт с первым пользователем рейтинг, а затем редактирует его для последующего вывода.
# Функция использует работу с файлом. 
def to_make_rating(nickname, scores):
    # Считываем в переменную все строки из файла rating.csv, где находятся баллы и имена.
    with open("rating.csv", "r", encoding = 'UTF-8') as r:
         lines = r.readlines()
    # Делаем проверку: если ник в файле и ник, который ввёл пользователь, одинаковые, 
    # то стираем всю старую информацию об этом нике, перезаписывая всю информацию в файл заново. 
    for i in lines:
        line = i.strip("\n").split(';')
        if line[0] == nickname:
            with open("rating.csv", "w", encoding = 'UTF-8' ) as f:
                  for line in lines:
                    if line.strip("\n").split(';')[0] != nickname:
                        f.write(line)
    # Независимо от работы цикла редактируем файл, добавляя введённый ник и баллы. 
    # Если раннее мы "стёрли" информацию о пользователе, мы добавим ее, но уже с новыми баллами. 
    with open('rating.csv','a', encoding = 'UTF-8') as r:
        print(nickname, scores, sep = ';', file = r)

# Функция to_show_rating выводит рейтинг из csv-файла. 
def to_show_rating():
    # Начинаем вывод рейтинга. 
    print('Рейтинг пользователей:')
    # Создаем в пустой словарь, где ключом будут баллы,
    # а значением - имя пользователя.
    dict_for_rating = {}  
    # Добавлям данные из csv-файла в словарь.
    with open('rating.csv', encoding = 'UTF-8') as r:
        for i in r:
            a = i.strip('\n').split(';')
            dict_for_rating[a[0]] = float(a[1])
    # Выводим рейтинг пользователей, отсортировав словарь по значениям и убыванию.
    for item in sort_dict_by_value(dict_for_rating, True):
        print(*item)

# Функция show_results выводит результаты(баллы и уровень),
# на среднем и продвинутом уровне результаты будут основаны на количестве замен.
## На начальном уровне результаты просто выводятся. 
@sampleDecorator
def show_results(counter, level, scores):
    print(f'Твои баллы: {scores}.')
    # Если пользователь получил 0 баллов на любом из уровней, попросим пользователя перепройти тест, 
    # так как весьма затруднительно определить его уровень, даже приблизительно. 
    if scores == 0:
        print('Сложно определить твой уровень. Может попробуешь ещё раз?')
    else:
        if level == 'elementary':
            print(f'Мы уверены, что у тебя начальный уровень(A) английского языка.')
        elif level == 'intermediate':
        # Если замен было 6 или больше, тогда уровень будет ниже.
            if counter >= 6:
                print(f'Мы думаем, что у тебя всё-таки начальный уровень(А) английского языка.')
            else:
                print(f'Мы уверены, что у тебя средний уровень(B) английского языка.')
        elif level == 'advanced':
        # Если замен было 6 или больше, тогда уровень будет ниже.
            if counter >= 6:
                print(f'Мы думаем, что у тебя всё-таки средний уровень(B) английского языка.')
            else:
                print(f'Мы уверены, что у тебя продвинутый уровень(C) английского языка.')

            
# Функция start_easy_test выводит вопросы начального уровня и считает баллы.
## Функция возвращает количество баллов. 
def start_easy_test(scores):
    for i in elementarylevel:
        # С помощью списка  alevel проверяем, какой вопрос будет выводиться пользователю,
        # если в i хранится вопрос, требующий multiple choice, выводим после вопроса варианты ответов,
        # если нет - просто выводим вопрос.
        if alevel[4] == i:
            print(f'{i} ')
            print(*a_level_answers, sep = '\n')
            answer = input(f'Твой ответ:').strip().upper()
        else:
            answer = input(f'{i} ').strip().upper()
        # Спрашиваем у пользователя вопрос и ожидаем ответ. 
        if answer == elementarylevel[i]: # Проверяем, совпадает ли ответ пользователя со значением ключа. 
            print('Правильный ответ!') 
            scores += 1  # За правильные ответы начисляется один балл. 
        else:
            print('Ответ неправильный.')
    return scores

# Функция start_interm_test выводит вопросы среднего уровня, 
# заменяет в определенных случаях вопросы и считает баллы и количество замен. 
## Функция возвращает количество баллов и замен вопросов. 
def start_interm_test(scores, counter):
    for i in intermlevel:
        # С помощью списка blevel проверяем, какой вопрос будет выводиться пользователю,
        # если в i хранится вопросы, требующие multiple choice, выводим после вопросов варианты ответов,
        # если нет - просто выводим вопрос.
        if blevel[1] == i: 
            print(f'{i} ')
            print(*b_level_answers[0:4], sep = '\n')
            answer = input(f'Твой ответ:').strip().upper()      
        elif blevel[7] == i:
            print(f'{i} ')
            print(*b_level_answers[4:], sep = '\n')
            answer = input(f'Твой ответ:').strip().upper()       
        else:
            answer = input(f'{i} ').strip().upper()
        # Спрашиваем у пользователя вопрос и ожидаем ответ.    
        if answer == intermlevel[i]:
            print('Правильный ответ!')
            scores += 1  
        # Предложим пользователю ответить на более легкий вопрос. 
        ## Используем функцию choice из модуля random. 
        ### Замена вопросов будет при каждом неправильном ответе.
        else: 
            counter += 1 # Если произошла замена, переменная будет увеличиваться.
            print('Давай попробуем вопрос полегче.')
            # В цикле работа схожа с функцией start_easy_test.  
            for i in range(1):
                q = random.choice(alevel)
                if alevel[4] == q:
                    print(f'{q} ')
                    print(*a_level_answers, sep = '\n')
                    answer1 = input(f'Твой ответ:').strip().upper()
                else:
                    answer1 = input(f'{q} ').strip().upper()
                if answer1 == elementarylevel[q]:  
                    print('Правильный ответ!')
                    scores += 0.5 # За правильный ответ в зоне "замены вопросов" дается 0.5 баллов. 
                else:
                    print('Ответ неправильный. Попыток больше нет.')
    return scores, counter 

# Функция start_adv_test выводит вопросы продвинутого уровня, 
# заменяет в определенных случаях вопросы и считает баллы и количество замен. 
## Принцип работы идентичен работе функции start_interm_test.
def start_adv_test(scores, counter):
    for i in advancedlevel:
        if clevel[1] == i:
            print(f'{i} ')
            print(*c_level_answers[0:4], sep = '\n')
            answer = input(f'Твой ответ:').strip().upper()
        elif clevel[3] == i:
            print(f'{i} ')
            print(*c_level_answers[4:8], sep = '\n')
            answer = input(f'Твой ответ:').strip().upper()
        elif clevel[9] == i:
            print(f'{i} ')
            print(*c_level_answers[8:], sep = '\n')
            answer = input(f'Твой ответ:').strip().upper()
        else:
            answer = input(f'{i} ').strip().upper()
        if answer == advancedlevel[i]:
            print('Правильный ответ!')
            scores += 1
        else:
            counter += 1
            print('Давай попробуем вопрос полегче.')
            for i in range(1):
                q = random.choice(blevel)
                if blevel[1] == q:
                    print(f'{q} ')
                    print(*b_level_answers[0:4], sep = '\n')
                    answer1 =input(f'Твой ответ:').strip().upper()
                elif blevel[7] == q:
                    print(f'{q} ')
                    print(*b_level_answers[4:], sep = '\n')
                    answer1 = input(f'Твой ответ:').strip().upper()
                else:
                    answer1 = input(f'{q} ').strip().upper()
                if answer1 == intermlevel[q]: 
                    print('Правильный ответ!')
                    scores += 0.5
                else:
                    print('Ответ неправильный. Попыток больше нет.')
    return scores, counter 

# ОСНОВНОЙ КОД

scores = 0 # В этой переменной будут считываяться баллы за правильные ответы. 
# Переменная counter считывает количество замен вопросов в тесте. 
## От её показателей будет зависеть точное определение уровня пользователя.
counter = 0
# Следующий шаг: представляемся пользователю.
print('Привет, я Сьюзи! Я помогу тебе узнать, какой у тебя уровень английского: начальный, средний или продвинутый:)')
# Начинаем тест и знакомим пользователя с правилами.
hello = input('Напиши "Начинаем", если хочешь начать прямо сейчас ').lower().strip()
if hello == 'начинаем':
    describe_rules()
    # Предлагаем пользователю выбрать уровень. 
    level = input('Выбери уровень сложности: elementary, intermediate или advanced. Введи соответствующий уровень: ').lower().strip()
    # Создаем цикл while, который работает пока не введут название уровня корректно.
    while level != 'elementary' and level != 'intermediate' and level != 'advanced':
        level = input('Не поняла тебя. Введи название уровня еще раз ').lower().strip()
    # В зависимости от уровня задаются вопросы из соответствующего словаря.
    if level == 'elementary':   
        scores = start_easy_test(scores)
        # Выводим результаты. 
        show_results(counter, level, scores)
        print('Удачи в изучении английского! До новых встреч!') # Прощаемся с пользователем.
    # Делаем то же самое при выборе других уровней. 
    elif level == 'intermediate':
        scores, counter = start_interm_test(scores, counter)
        show_results(counter, level, scores)
        print('Удачи в изучении английского! До новых встреч!') 
    # Совершаем со сложным уровнем те же действия, что и со средним. 
    elif level == 'advanced':
        scores, counter = start_adv_test(scores, counter)
        show_results(counter, level, scores)
        # После прохождения только продвинутого уровня будет предложено быть в рейтинге. 
        ## Запрашиваем у пользователя имя для включения его в рейтинг
        ## или слово "нет", если пользователь этого не хочет.
        nickname = input('Напиши свой ник со знаком @, если хочешь быть в рейтинге. Напиши "нет", если не хочешь. ').lower().strip()
        if nickname != 'нет':
            # Создаём цикл, пока пользователь не ведёт ник правильно. 
            while '@' not in nickname:
                nickname = input('Введи ник ещё раз, со знаком "@":')
            # Записываем информацию в файл. 
            rating = to_make_rating(nickname, scores)
            # Выводим рейтинг пользователей.
            rating = to_show_rating()
            print('До новых встреч!')
        else:
            # Если пользователь не хочет быть в рейтинге, 
            # выведем ему рейтинг других участников. 
            print('Хорошо, я тебя поняла.')
            rating = to_show_rating()
            print('Удачи в усовершенствовании английского! До новых встреч!')      
else:
    print('Не хочешь проходить тест? Тогда я жду тебя в другое время!')
```
